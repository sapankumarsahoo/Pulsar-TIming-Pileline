```python
# Install psrqpy if not already installed
# !pip install psrqpy

import psrqpy
import matplotlib.pyplot as plt

# Query the ATNF Pulsar Catalogue for the required data
query = psrqpy.QueryATNF(params=['P0', 'DM', 'P1', 'BSURF', 'EDOT', 'AGE'])
pulsar_data = query.pandas

# Remove rows with missing data for any of the required parameters
pulsar_data = pulsar_data.dropna(subset=['P0', 'DM', 'P1', 'BSURF', 'EDOT', 'AGE'])

# Filter data for DM between 0 and 300
normal_pulsar = pulsar_data[(pulsar_data['DM'] >= 0) & (pulsar_data['DM'] <= 300)]

# Split the data into NPs and normal pulsars (NPs: P0 >= 30 ms)
normal_pulsar = normal_pulsar[normal_pulsar['P0'] >= 0.03]
total_NP = len(normal_pulsar)



# Create subplots to plot DM vs multiple parameters
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# DM vs Period Derivative (P1)

pulsars_in_dm_range_period_NP = normal_pulsar[(normal_pulsar['DM'] <= 250) & (normal_pulsar['P1'] <= 0.4e-13) & (normal_pulsar['P1'] >= 1e-16)]
fraction_of_pulsars1_NP = len(pulsars_in_dm_range_period_NP) / total_NP



axes[0, 0].scatter(normal_pulsar['DM'], normal_pulsar['P1'], c='blue', alpha=0.5)
axes[0, 0].set_xlabel('Dispersion Measure (pc cm⁻³)')
axes[0, 0].set_ylabel('Period Derivative (s/s)')
axes[0, 0].set_title('DM vs Period Derivative')
axes[0, 0].set_ylim([1e-16,0.4e-13])  # Set Y-axis range
axes[0, 0].axvline(x=250, color='green', linestyle='--', linewidth=2)  # Vertical line at DM = 250
axes[0, 0].text(75, 1e-14, f'Fraction of NP (P_dot ~10^-16 to 0.4e-13, DM <= 250 ): {fraction_of_pulsars1_NP:.2f}', color='black', fontsize=14, alpha=0.9)


# DM vs Surface Magnetic Field (Bsurf)
pulsars_in_dm_range_B = normal_pulsar[(normal_pulsar['DM'] <= 250) & (normal_pulsar['BSURF'] >= 1e11) & (normal_pulsar['BSURF'] <= 1e13)]
fraction_of_pulsars2 = len(pulsars_in_dm_range_B) / total_NP



axes[0, 1].scatter(normal_pulsar['DM'], normal_pulsar['BSURF'] / 1e11, c='blue', alpha=0.5)
axes[0, 1].set_xlabel('Dispersion Measure (pc cm⁻³)')
axes[0, 1].set_ylabel('Surface Magnetic Field (x10^8 Gauss)')
axes[0, 1].set_title('DM vs Surface Magnetic Field')
axes[0, 1].set_ylim([1, 100])  # Set Y-axis range
axes[0, 1].axvline(x=250, color='green', linestyle='--', linewidth=2)  # Vertical line at DM = 250
axes[0, 1].text(80, 8, f'Fraction of Pulsars (B ~ 10^8 to 10^9 Gauss/s, DM <= 250): {fraction_of_pulsars2:.2f}', color='black', fontsize=14, alpha=0.9)

# DM vs Spin-down Luminosity (Edot)
pulsars_in_dm_range_EDOT = normal_pulsar[(normal_pulsar['DM'] <= 250) & (normal_pulsar['EDOT'] >= 1e30) & (normal_pulsar['EDOT'] <= 1e34)]
fraction_of_pulsars3 = len(pulsars_in_dm_range_EDOT) / total_NP
axes[1, 0].scatter(normal_pulsar['DM'], normal_pulsar['EDOT'] / 1e30, c='blue', alpha=0.5)
axes[1, 0].set_xlabel('Dispersion Measure (pc cm⁻³)')
axes[1, 0].set_ylabel('Spin-down Luminosity (x10^32 erg/s)')
axes[1, 0].set_title('DM vs Spin-down Luminosity')
axes[1, 0].set_ylim([1, 10000])  # Set Y-axis range
axes[1, 0].axvline(x=250, color='green', linestyle='--', linewidth=2)  # Vertical line at DM = 250
axes[1, 0].text(80, 300, f'Fraction of NPs (E_dot ~ 10^32 to 4*10^35, DM <= 250): {fraction_of_pulsars3:.2f}', color='black',fontsize=14, alpha=0.9)


# DM vs Characteristic Age
pulsars_in_dm_range_AGE = normal_pulsar[(normal_pulsar['DM'] <= 250) & (normal_pulsar['AGE'] >= 1e3) & (normal_pulsar['AGE'] < 1e8)]
fraction_of_pulsars4 = len(pulsars_in_dm_range_AGE) / total_NP



axes[1, 1].scatter(normal_pulsar['DM'], normal_pulsar['AGE'] / 1e4, c='blue', alpha=0.5)
axes[1, 1].set_xlabel('Dispersion Measure (pc cm⁻³)')
axes[1, 1].set_ylabel('Characteristic Age (x10⁹ years)')
axes[1, 1].set_title('DM vs Characteristic Age')
axes[1, 1].set_ylim([1, 1000])  # Set Y-axis range
axes[1, 1].axvline(x=250, color='green', linestyle='--', linewidth=2, alpha=1)  # Vertical line at DM = 250
axes[1, 1].text(80, 8, f'Fraction of NPs (Age ~ 10^3 to 10^8, DM <= 250: {fraction_of_pulsars4:.2f}', color='black', fontsize=14, alpha=0.9)

# Set x-axis limits to [0, 300] for all plots
for ax in axes.flat:
    ax.set_xlim([0, 300])

# Add a legend
axes[0, 0].legend(loc='upper right')
axes[0, 1].legend(loc='upper right')
axes[1, 0].legend(loc='upper right')
axes[1, 1].legend(loc='upper right')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```
