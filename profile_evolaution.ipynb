```python
import numpy as np
import matplotlib.pyplot as pltimport numpy as np
from scipy.signal import find_peaks
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

file = np.loadtxt('700_100_MHz.txt')
x = file[:, 0]
y_original = file[:, 1]
y = y_original + np.max(y_original)
peaks, _ = find_peaks(y, height=0.9)  # Adjust threshold as needed

# Define Gaussian function
def gaussian(x, *params):
    A = params[::3]
    mu = params[1::3]
    sigma = params[2::3]
    result = np.zeros_like(x)
    for i in range(len(A)):
        result += A[i] * np.exp(-(x - mu[i])**2 / (2 * sigma[i]**2))
    return result

def fwhm(x, y):
    half_max = max(y) / 2
    pos = np.where(y >= half_max)[0]
    fwhm_value = x[pos[-1]] - x[pos[0]]
    return fwhm_value

def plot_peaks(x, y, peaks, params=None):
    plt.plot(x, y, label='Profile', color='black')
    if params is not None:
        plt.plot(x, gaussian(x, *params), label='Multi-component Gaussian Fit', linestyle='dashed', color='r')
        for i, peak in enumerate(peaks):
            # Calculate FWHM for each peak
            peak_height = params[3*i]
            peak_center = params[3*i + 1]
            peak_sigma = params[3*i + 2]
            fwhm_value = fwhm(x, gaussian(x, *params))
            print(f"FWHM for Peak {i+1}: {fwhm_value}")

    plt.plot(x[peaks], y[peaks], "x", label="Peaks")
    plt.legend()
    plt.title("Profile with Multi-component Gaussian Fit to Peaks")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()

# Initial plot with peaks and Gaussian fit
plot_peaks(x, y, peaks)

# Interactive deletion of peaks
delete_peak = True
while delete_peak:
    delete_index = input("Enter the index of the peak to delete (0 to keep all peaks): ")
    if delete_index == '0':
        delete_peak = False
    else:
        delete_index = int(delete_index)
        peaks = np.delete(peaks, delete_index - 1)
        plot_peaks(x, y, peaks)

# Fit Gaussian to remaining peaks
initial_guess = []
for peak in peaks:
    initial_guess.extend([y[peak], x[peak], 1.0])

popt, _ = curve_fit(gaussian, x, y, p0=initial_guess)

# Final plot with remaining peaks and Gaussian fit
plot_peaks(x, y, peaks, popt)


# Load data from the file
file1 = np.loadtxt('400_MHz.txt')
intensity1 = file1[:,1]# Display the plot after all lines are drawn

# Display the plot after all lines are drawn
phase3 = file1[:,0]
# Load data from the file
file2 = np.loadtxt('700_100_MHz.txt')
intensity2 = file2[:,1]
phase4 = file2[:,0]

modified_intensity1 = intensity1 + 16000
modified_intensity2 = intensity2 + 16000

normalized_intensity1 = modified_intensity1 / np.max(modified_intensity1)
normalized_intensity2 = modified_intensity2 / np.max(modified_intensity2)

# Plot each line for different indices
#for i in range(32):
plt.figure(figsize=(10, 6))

plt.plot(phase4/32,normalized_intensity2 + 1, label='650 MHz', color='b')
plt.plot(phase3/32,normalized_intensity1, color= 'r', label='400 MHz')
#plt.axvline(x=0.562, color='g', linestyle='--', alpha = 1)  # Adjusted x value for demonstration
#plt.axvline(x=0.155,color='g',linestyle='--',alpha = 1 )
plt.title('Profile Evolution of J0248+4230',fontsize = 25)
plt.ylabel('Normalized Flux Density', fontsize=16)
plt.xlabel('Phase', fontsize=16)
plt.legend()
plt.show()
##J0248+4230_pa_500_200_4096_4_1_8_31may2021.raw.64_PSR_0248+4230.ar _ 350MHz
#J0248+4230_pa_550_200_4096_4_1_8_05mar2018.raw.gptool_PSR_0248+4230.pfd.ps_700MHz



```
