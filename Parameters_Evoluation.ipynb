```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming your data is in a CSV file named 'pulsar_data.csv'
file_path = '/home/ssahoo/Desktop/Project_2/code/Parameters_evolution.csv'

# Read data from the CSV file using pandas
data = pd.read_csv(file_path, delim_whitespace=True)  # Assuming the data is space-delimited

# Sort the data based on the 'dates' column
data.sort_values(by='dates', inplace=True)

# Extract columns
dates = pd.to_datetime(data['dates'])
periods = data['periods']
errors = 0# data['errors']  # Assuming errors are zero for simplicity

# Plot the graph with both y-axis and error bars in log scale
plt.figure(figsize=(10, 6))
#plt.yscale('log')  # Set y-axis to log scale
plt.errorbar(dates, periods, yerr=errors, fmt='o-', label='Pulsar Periods')
plt.xlabel('Dates')
plt.ylabel('Change in Pulsar Periods (log scale)')

plt.title('Variation in Pulsar Periods over Time ')
plt.legend()

# Remove Y-axis tick labels
plt.yticks([])

# Show the plot
plt.show()
```
