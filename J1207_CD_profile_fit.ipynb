```python
from scipy.optimize import curve_fit
from scipy.special import erfc

# Define the Gaussian convolved with exponential function
def gauss_exp_convolve(x, sigma, tau, amp):
    """
    Function to model Gaussian convolved with exponential.
    sigma: standard deviation of Gaussian
    tau: scattering timescale (from exponential decay)
    amp: amplitude of the function
    """
    return (amp / (2 * tau)) * np.exp((sigma**2) / (2 * tau**2) - (x / tau)) * erfc((sigma / (np.sqrt(2) * tau)) - (x / (np.sqrt(2) * sigma)))

# Set initial guesses for sigma, tau, and amp (amplitude)
initial_guesses = [5, 5, 1]

# Fit the function to the 128-bin data
popt_128, _ = curve_fit(gauss_exp_convolve, x_128_bins, array_128_bins_normalized, p0=initial_guesses, maxfev=100000000)

# Fit the function to the 64-bin data (aligned)
popt_64, _ = curve_fit(gauss_exp_convolve, x_64_bins_shifted, array_64_bins_normalized, p0=initial_guesses, maxfev=100000000)

# Extract the scattering timescales (tau) from both fits
tau_128 = popt_128[1]
tau_64 = popt_64[1]

# Plotting the results for the 128-bin data
plt.figure(figsize=(10, 6))
plt.plot(x_128_bins, array_128_bins_normalized, label='128 Bins Data', color='blue', marker='o')
plt.plot(x_128_bins, gauss_exp_convolve(x_128_bins, *popt_128), label=f'Fit (tau={tau_128:.2f})', color='green')
plt.title('128 Bins Pulsar Profile with Gaussian-Exponential Fit')
plt.xlabel('Bin Number')
plt.ylabel('Normalized Intensity')
plt.legend()
plt.show()

# Plotting the results for the 64-bin data
plt.figure(figsize=(10, 6))
plt.plot(x_64_bins_shifted, array_64_bins_normalized, label='64 Bins Data', color='red', marker='x')
plt.plot(x_64_bins_shifted, gauss_exp_convolve(x_64_bins_shifted, *popt_64), label=f'Fit (tau={tau_64:.2f})', color='purple')
plt.title('64 Bins Pulsar Profile with Gaussian-Exponential Fit')
plt.xlabel('Bin Number')
plt.ylabel('Normalized Intensity')
plt.legend()
plt.show()

# Display the calculated scattering timescales
tau_128, tau_64
```

```python
# Import required libraries
import psrqpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Query the ATNF Pulsar Catalog using psrqpy
# We are interested in the pulsar's Galactic latitude and distance
query = psrqpy.QueryATNF(params=["PSRJ", "DIST", "GB"]).pandas

# Remove pulsars with missing distance or Galactic latitude
pulsars_with_distance_and_GB = query[(~query['DIST'].isnull()) & (~query['GB'].isnull())]

# Convert Galactic latitude from degrees to radians for the calculation
pulsars_with_distance_and_GB['GB_rad'] = np.radians(pulsars_with_distance_and_GB['GB'])

# Calculate the z-component (height above/below the Galactic plane) in parsecs
pulsars_with_distance_and_GB['z'] = pulsars_with_distance_and_GB['DIST'] * 1000 * np.sin(pulsars_with_distance_and_GB['GB_rad'])

# Constants for the Galactic potential model
a_0 = 1.5e-9  # km/s^2, arbitrary constant for acceleration
z_0 = 500  # pc, scale height of the Galactic disk

# Calculate line-of-sight acceleration (a_z) using the given formula
pulsars_with_distance_and_GB['a_z'] = -a_0 * pulsars_with_distance_and_GB['z'] / np.sqrt(pulsars_with_distance_and_GB['z']**2 + z_0**2)

# Bin the pulsars by their z values to count the number of pulsars at different heights
z_bins = np.linspace(-3000, 3000, 3000)  # 30 bins between -3000 pc and 3000 pc
pulsars_with_distance_and_GB['z_bin'] = np.digitize(pulsars_with_distance_and_GB['z'], z_bins)

# Calculate the number of pulsars in each z-bin
pulsar_counts = pulsars_with_distance_and_GB.groupby('z_bin').size()

# Calculate the average a_z for each bin
average_a_z_per_bin = pulsars_with_distance_and_GB.groupby('z_bin')['a_z'].mean()

# Remove any bins with no pulsars
pulsar_counts = pulsar_counts[pulsar_counts > 0]
average_a_z_per_bin = average_a_z_per_bin[pulsar_counts.index]

# Plot the line-of-sight acceleration vs the number of pulsars in terms of z
plt.figure(figsize=(10, 6))
plt.scatter(average_a_z_per_bin, pulsar_counts, color='blue', label='Pulsars')
plt.title('Line-of-Sight Acceleration vs Number of Pulsars (in terms of z)')
plt.xlabel('Line-of-Sight Acceleration (a_z in km/s^2)')
plt.ylabel('Number of Pulsars')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Import required libraries
import psrqpy
import numpy as np
import matplotlib.pyplot as plt

# Query the ATNF Pulsar Catalog using psrqpy
# We will fetch P0, P1 (period and its derivative), DIST (distance), and GL (Galactic Latitude)
query = psrqpy.QueryATNF(params=["PSRJ", "P0", "P1", "DIST", "GL", "GB"]).pandas

# Filter out pulsars with missing distance and proper motion data
pulsars_with_data = query[~query['DIST'].isnull() & ~query['P0'].isnull() & ~query['P1'].isnull() & ~query['GB'].isnull()]

# Classify pulsars: MSPs typically have a period less than 30 ms (0.03 s)
MSPs = pulsars_with_data[pulsars_with_data['P0'] < 0.03]
normal_pulsars = pulsars_with_data[pulsars_with_data['P0'] >= 0.03]

# Calculate Galactic height (z) for both MSPs and normal pulsars
# z = d * sin(b), where d is the distance and b is the Galactic latitude (GB)
MSPs['z'] = MSPs['DIST'] * np.sin(np.radians(MSPs['GB']))  # Convert GB to radians
normal_pulsars['z'] = normal_pulsars['DIST'] * np.sin(np.radians(normal_pulsars['GB']))

# Calculate line-of-sight acceleration using period second derivative P1 (approximation)
# Line-of-sight acceleration = (P1 / P0)
MSPs['acceleration'] = MSPs['P1'] / MSPs['P0']
normal_pulsars['acceleration'] = normal_pulsars['P1'] / normal_pulsars['P0']

# Bin the data based on Galactic height (z)
z_bins = np.linspace(-1.5, 1.5, 30)  # Adjust as needed for the z range in kpc
MSP_binned, _ = np.histogram(MSPs['z'], bins=z_bins)
normal_binned, _ = np.histogram(normal_pulsars['z'], bins=z_bins)

# Plot line-of-sight acceleration vs number of pulsars (for MSPs and normal pulsars)
plt.figure(figsize=(10, 6))

# MSPs: plot their acceleration in terms of z
plt.scatter(MSPs['z'], MSPs['acceleration'], color='blue', label='MSPs', alpha=0.7)

# Normal Pulsars: plot their acceleration in terms of z
plt.scatter(normal_pulsars['z'], normal_pulsars['acceleration'], color='green', label='Normal Pulsars', alpha=0.7)

# Customize plot
plt.title('Line-of-Sight Acceleration vs Galactic Height (z)')
plt.xlabel('Galactic Height (z) [kpc]')
plt.ylabel('Line-of-Sight Acceleration [s/s]')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.legend()

# Show the plot
plt.show()
```

```python
import psrqpy
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

# Query the ATNF pulsar catalog using psrqpy
query = psrqpy.QueryATNF(params=['P0', 'P1', 'DIST', 'GL', 'GB', 'AGE'])
catalog = query.get_pulsars()

# Extract relevant data
galactic_longitudes = np.array(catalog['GL'], dtype=np.float32)  # Galactic Longitude (l)
galactic_latitudes = np.array(catalog['GB'], dtype=np.float32)  # Galactic Latitude (b)
distances = np.array(catalog['DIST'], dtype=np.float32)  # Distance in kpc
periods = np.array(catalog['P0'], dtype=np.float32)  # Spin periods
period_derivatives = np.array(catalog['P1'], dtype=np.float32)  # Period derivative
ages = np.array(catalog['AGE'], dtype=np.float32)  # Pulsar ages

# Filter out NaN values from data
valid_data_mask = (~np.isnan(galactic_longitudes)) & (~np.isnan(galactic_latitudes)) & (~np.isnan(distances))
galactic_longitudes = galactic_longitudes[valid_data_mask]
galactic_latitudes = galactic_latitudes[valid_data_mask]
distances = distances[valid_data_mask]

# Convert distances from kpc to parsecs for consistency (1 kpc = 1000 parsecs)
distances *= 1000

# Convert Galactic coordinates to Cartesian (x, y, z) coordinates
def galactic_to_cartesian(l, b, d):
    l_rad = np.radians(l)
    b_rad = np.radians(b)
    x = d * np.cos(b_rad) * np.cos(l_rad)
    y = d * np.cos(b_rad) * np.sin(l_rad)
    z = d * np.sin(b_rad)
    return x, y, z

x, y, z = galactic_to_cartesian(galactic_longitudes, galactic_latitudes, distances)

# ----------------------------- Visualization ---------------------------- #

# 1. Pulsar Distribution in the Galactic Plane (2D plot)
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=10, alpha=0.6, color='blue')
plt.title("Pulsar Distribution in the Galactic Plane (x-y plane)")
plt.xlabel('X (Parsecs)')
plt.ylabel('Y (Parsecs)')
plt.grid(True)
plt.show()

# 2. 3D Pulsar Distribution
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, s=10, alpha=0.6, color='red')
ax.set_title("3D Pulsar Distribution in the Milky Way")
ax.set_xlabel('X (Parsecs)')
ax.set_ylabel('Y (Parsecs)')
ax.set_zlabel('Z (Parsecs)')
plt.show()

# 3. Histogram of Pulsar Positions (Latitude Distribution)
plt.figure(figsize=(10, 6))
sns.histplot(galactic_latitudes, bins=30, kde=True, color='green')
plt.title("Distribution of Pulsars by Galactic Latitude (b)")
plt.xlabel('Galactic Latitude (b)')
plt.ylabel('Number of Pulsars')
plt.grid(True)
plt.show()

# 4. Radial Distribution (Distance from Galactic Center)
radial_distances = np.sqrt(x**2 + y**2)
plt.figure(figsize=(10, 6))
sns.histplot(radial_distances, bins=30, kde=True, color='purple')
plt.title("Radial Distribution of Pulsars (Distance from Galactic Center)")
plt.xlabel('Radial Distance (Parsecs)')
plt.ylabel('Number of Pulsars')
plt.grid(True)
plt.show()

# 5. Distribution of Pulsar Ages
plt.figure(figsize=(10, 6))
valid_ages = ages[~np.isnan(ages)]  # Remove NaN values from age data
sns.histplot(valid_ages, bins=30, kde=True, color='orange')
plt.title("Distribution of Pulsar Ages")
plt.xlabel('Age (years)')
plt.ylabel('Number of Pulsars')
plt.grid(True)
plt.show()

# 6. Pulsar Height Distribution Above and Below the Galactic Plane (z-axis)
plt.figure(figsize=(10, 6))
sns.histplot(z, bins=30, kde=True, color='blue')
plt.title("Distribution of Pulsars Above and Below Galactic Plane (Z-axis)")
plt.xlabel('Height (Parsecs)')
plt.ylabel('Number of Pulsars')
plt.grid(True)
plt.show()
```

```python
import psrqpy
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import data from the ATNF catalog
# Use psrqpy to query the catalog for pulsar data
query = psrqpy.QueryATNF()
# Fetch data; you can adjust parameters based on your needs
pulsars = query.query_pulsars()

# Step 2: Convert the data to a DataFrame for easier analysis
pulsar_df = pulsars.to_dataframe()

# Step 3: Filter the DataFrame for relevant columns
# For spectral analysis, we may be interested in the following columns:
# - 'PSR' (Pulsar Name)
# - 'SpectralIndex' (Spectral index)
# - 'Luminosity' (Luminosity)
# - 'DM' (Dispersion Measure)

# Ensure to adjust column names based on the actual DataFrame structure
spectral_df = pulsar_df[['PSR', 'SpectralIndex', 'Luminosity', 'DM']].dropna()

# Step 4: Perform basic statistical analysis
# Describe the data
print(spectral_df.describe())

# Step 5: Visualization of the Spectral Index
plt.figure(figsize=(10, 6))
plt.hist(spectral_df['SpectralIndex'], bins=30, color='blue', alpha=0.7)
plt.title('Distribution of Spectral Indices of Pulsars')
plt.xlabel('Spectral Index')
plt.ylabel('Number of Pulsars')
plt.grid()
plt.show()

# Step 6: Scatter plot of Luminosity vs Spectral Index
plt.figure(figsize=(10, 6))
plt.scatter(spectral_df['SpectralIndex'], spectral_df['Luminosity'], alpha=0.5)
plt.title('Luminosity vs Spectral Index of Pulsars')
plt.xlabel('Spectral Index')
plt.ylabel('Luminosity')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the period range from 1 ms to 1 sec (in seconds)
periods = np.linspace(1, 100, 1000)  # 100 values from 1 ms to 1 sec

# Calculate tolerance for each period
tolerances = 0.1*(3+ ((periods) * (60 / 999)+ 1) * (0.1/0.1))

# Plotting
plt.figure(figsize=(20, 15))
plt.plot(periods, tolerances, label='Tolerance', color='blue')
plt.title('Tolerance vs Period')
plt.xlabel('Period (Milli seconds)')
plt.ylabel('Tolerance')
plt.grid()
plt.legend()
plt.xscale('linear')
plt.yscale('linear')
plt.xticks(np.arange(0, 100, 1), rotation=45)
plt.yticks(np.arange(0, max(tolerances) , max(tolerances)/10))
plt.tight_layout()
plt.show()
```
