```python
import numpy as np
file_path = 'SNR_time.txt'
data = np.loadtxt(file_path, skiprows=1)



SNR2 =5 #5 sigma deteciton
G=0.32 *13
T_sys= 106 #TT_rec+T_sky: kelvin --> Band3 (Best RMS sensitivity 0.04 mJy with all antennas) (T = 93 ), band4 (T =60+10=70) (Best RMS sensitivity 0.02 mJy)
B=200e6 #Hz
S_sys = T_sys / G
n=20 #no of antenna
P=0.002600999083 #sec
W= 0.002600999083 * 0.41 #sec (BasicallyTaking 10% of the time period, here I have calculate it)
total_bin_number = 32
on_bin_number = 12 #40% due to wider profile
#Best rms sensitivity achieved in band-3 is 0.0.mJy

SNR =data[:,0]
print(f"Total Number of data points:: {len(SNR)}")
t=data[:,1] #sec

#Calculation od S_pulsar (SNR = 5 for detection)
def S_mean(SNR2,T_sys,W,G,n,t,B,P):
    S_mean = SNR2 * 1e3 * T_sys * (W**0.5) / ( G * ((t*n*B)**0.5) * ((P-W)**0.5) )  #converted into Jy to SI units and in units of mJy
    return S_mean
s = S_mean(SNR2,T_sys,W,G,n,t,B,P)
print("The S_mean value is (for 5 sigma detection) is::" , S_mean(SNR2,T_sys,W,G,n,t,B,P),"mJy  \n S_means = Pulsar_total_flux_density / on-pulse_beam bin number")
def S_pulsar(SNR,T_sys,W,G,n,t,B,P):
    S_pulsar = SNR * 1e3 * T_sys * (W**0.5) / ( G * ((t*n*B)**0.5) * ((P-W)**0.5) )  #converted into Jy to SI units and in units of mJy
    return S_pulsar
S = S_pulsar(SNR2,T_sys,W,G,n,t,B,P)


total_S_pulsar = S_pulsar(SNR,T_sys,W,G,n,t,B,P) * on_bin_number


print("The S_mean value for the pulsar is::" , S_pulsar(SNR,T_sys,W,G,n,t,B,P),"mJy")
print(f"Maximum S_mean = {np.max(S_pulsar(SNR,T_sys,W,G,n,t,B,P))} mJy")
print(f"The pulsar flux density is = S_mean * phase_bin-on-pulsed-region, and is given by {total_S_pulsar} mJy")
print(f'The average value of pulsar flux density is :{np.average(total_S_pulsar)} mJy')
print(f"Maximum Values of pulsar flux density is :: {np.max(total_S_pulsar)}")

def TOA(S_sys,P,B,t,S_mean):
    TOA = 10**6 * S_sys * P * (W/P)**1.5 / ( (t*B)**0.5 * S_mean )  #divided with 13**0.5 for array of 13 antenna.# in microseconds units
    return TOA
#rms_toa =  TOA(S_sys,P,B,t,S_mean)
print("The RMS_TOA value is",TOA(S_sys,P,B,t,s), "in micro seconds")
```
